```python
# Import the modules that we will need for plotting
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
import scipy.stats as stats
```

```python
# Make some data
x   =   np.arange(0,10)
y   =   x+np.random.rand(10)*3.
print(x)
print(y)
```

```python
# Plot the data
plt.close('all') # to close all open windows
fig = plt.figure(figsize=(6,4))
ax1 = fig.add_subplot(111)
ax1.plot(x, y, 'x', mew=2, ms=8, color='blue')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
```

```python
# Fit a straight line
slope,intercept,r_value,p_value,std_err     =   stats.linregress(x,y)
print(slope,intercept)
```

```python
# Plot result
ax1.plot(x,x*slope+intercept,ls='--',color='red',lw=2)
```

```python
print('r^2: %s' % r_value)
print('p-value: %s' % p_value)
```
