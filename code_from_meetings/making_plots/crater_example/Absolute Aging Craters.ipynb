## Background Info

* There are 2 different ways of aging craters: relative and absolute aging.
* Several Apollo and Luna missions to the Moon have returned Moon rocks which have been radioactively dated to be up to 4.44 billion years old.
* Because of the Apollo and Luna Moon rocks the oldest absolutely aged crater is 4.44 billion years old.
* Different crater characteristics put crater ages into certain groupings of times:
        Late Heavy Bombardment craters age up to ~3.9 Gy
        Terrae Highlands craters age up to ~4.5 Ga
        Mare craters ages 3.1-3.8 Ga
        Terrae Basin craters ages 3.8-4.0 Ga   
* At some point the surface is covered with craters and crater saturation occurs. Crater saturation makes absolutely aging craters very difficult without analyzing a sample of Moon rock from the same area.

### Lunar Geoligical Time Scale

* The Lunar Time Scale is seperated into 5 different time periods.
    1. Pre-Nectarian 3.92 to 4.6 by
    2. Nectarian 3.85 to 3.92 by
    3. Imbrian 3.2 to 3.85 by
    4. Eratosthenian 1.0 to 3.2 by
    5. Copernican 0 to 1.0 by

### Fixing Problems That Arose Within Our Data Set Due To Typing Errors

### The grouby function arranges all absolutely aged craters into catagories.

```python
group_df = fixedages.groupby('Fixed_Age')
count = group_df.count()
count.columns = ['Count']
count
```

The majority of impacts happened in the Nectarian period! (3,920-3,850 million years ago)

### The answer as to why some craters are absolutely aged and some cannot be is related to crater saturation.  Craters can be aged as long as lunar samples have been analyzed and crater saturation has not occured.

* Crater saturation happens when 100% of the surface has been covered with craters.  This makes it difficult to age craters according to superposition because the oldest craters are now bombarded with newer ones making the older craters much harder to see.

### Getting image dimensions in order to plot an ellipse.  The area is in pixels.

```python
from PIL import Image, ImageDraw

def get_num_pixels(crater_saturation):
    width, height = Image.open(crater_saturation).size
    return width, height

get_num_pixels("./Crater Saturation.jpg")

#Lunar Surface, gets the image pixel dimensions to plot ellipses
```

### ImageDraw is a tool in PIL that allows geometric shapes to be drawn on images.

```python
im = Image.open("./Crater Saturation.jpg")

bbox =  (140,210,268,275) #The trickiest thing was to realize the origin is actually the top left side of the image.
#The dimensions are of the bounding box(bbox) where an ellipse will be drawn inside of the box.  The first two
#coordinates are for the top left corner of the bounding box and the last two coordinates are for the bottom right
#corners. The ellipse will only be seen if the coordinates make a complete rectangle or square.
draw = ImageDraw.Draw(im)
draw.ellipse(bbox,outline=(255,0,0)) #(255,0,0) is an RGB representation of what color the ellipse should be.
del draw

im
```

This is a good example of superposition and how many smaller craters can eradicate any existence of there being a bigger and older crater underneath.  There could also be an impact making a crater size much larger than one below it.  Because of crater saturation it is difficult to tell the absolute ages of craters based on superposition.

### Same ImageDraw use for an ellipse as before.  Just showing a different part of the image for more crater saturation clarification.

```python
im = Image.open("./Crater Saturation.jpg")

bbox =  (0,0,275,75)
draw = ImageDraw.Draw(im)
draw.ellipse(bbox,outline=(255,0,0))
del draw

im
```

This is another example of crater saturation.  The top left surface could be the reminant of a massive crater floor, but it is hard to tell because later craters have wiped out the crater walls of the earlier formed crater.
